```python
import pandas as pd
df = pd.read_csv('data.csv')
df.head()
```

```python
readmit_values = df['prev_readmissions'].value_counts()
readmit_means = df.groupby('prev_readmissions')['label'].mean()

def readmit_groups(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return 2
    
groups = df['prev_readmissions'].map(readmit_groups)
df['prev_readmit_group'] = groups
```

```python
df.groupby('comorbidities_count')['label'].agg(['count', 'mean'])

def comorbidities_groups(x):
    if x == 1:
        return 'one'
    elif x == 2:
        return 'two'
    elif x == 3:
        return 'three'
    elif x == 4:
        return 'four'
    else:
        return 'five or more'

groups = df['comorbidities_count'].map(comorbidities_groups)
df['comorbidities_group'] = groups
```

```python
df.groupby('length_of_stay')['label'].agg(['count', 'mean'])
```

```python
df['los_group'] = df['length_of_stay'].map({
    3:'3 to 4',
    4: '3 to 4',
    5: '5',
    6: '6 to 8',
    7: '6 to 8',
    8: '6 to 8',
    }).fillna('9+')
df.groupby('los_group')['label'].agg(['count', 'mean'])

# 6, 7, 8
```

```python
df['medications_group'] = df['medications_count'].replace({
    **{k: '8 to 14' for k in [8, 9, 10, 11, 12, 13, 14]},
    **{j: '15+' for j in [15, 16, 17, 18]}
})

df.groupby('medications_group')['label'].agg(['count', 'mean'])
```

```python
df.groupby('followup_visits_last_year')['label'].agg(['count', 'mean'])
```

```python
df['followup_visits_group'] = df['followup_visits_last_year'].replace({
    0: '0 to 1',
    1: '0 to 1',
    2: '2',
    3: '3',
    4: '4',
    5: '5 or more',
    6: '5 or more',
    7: '5 or more',
    8: '5 or more',
    9: '5 or more',
    10: '5 or more',
})
```

```python
agg = df.groupby('followup_visits_group')['label'].agg(['count', 'mean'])
agg
```

```python
df['admission_date'] = pd.to_datetime(df['admission_date'], errors='coerce')
df['admission_month'] = df['admission_date'].dt.month
df['admission_year'] = df['admission_date'].dt.year
```

```python
df.groupby('admission_year')['label'].agg(['count', 'mean'])
```

```python
months = df.groupby('admission_month')['label'].agg(['count', 'mean'])
months.describe()
```

```python
means = df.groupby('admission_month')['label'].mean()
means
```

```python
df['admission_day'] = df['admission_date'].dt.day
overview = df.groupby('admission_day')['label'].agg(['count','mean'])
overview.describe()
overview.mean()
```

```python
overall = df['label'].mean()
lower = overall - 0.05
upper = overall + 0.05
day_means = df.groupby('admission_day')['label'].mean()
```

```python
day_means[day_means <= lower]
```

```python
high_outlier = day_means[day_means >= upper]
high_outlier
```

```python
high_outlier.value_counts()
```

```python
df['days_of_week'] = df['admission_date'].dt.day_name()
days_overview = df.groupby('days_of_week')['label'].mean()
days_overview.describe()
```

```python
overall = df['label'].mean()
lower = overall - 0.05
upper = overall + 0.05
low_outliers = days_overview[days_overview <= lower]
high_outliers = days_overview[days_overview >= upper]
```

```python
print('low:')
print(low_outliers)
print('high')
print(high_outliers)
```
