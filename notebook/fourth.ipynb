```python
import pandas as pd
df = pd.read_pickle('cleaned_two.pkl')
df.head()
```

```python
df.groupby('discharge_disposition')['label'].agg(['count', 'mean'])
```

```python
df['dc_location'] = df['discharge_disposition']
df['dc_location'] = df['dc_location'].replace({
    'Home': 'Home/Rehab',
    'Rehab': 'Home/Rehab'
})
```

```python
df['dc_location'].value_counts()
```

```python
df.groupby('primary_diagnosis')['label'].agg(['count', 'mean'])
```

```python
# primary dx tiers:
means = df.groupby('primary_diagnosis')['label'].mean()
def tier(mean):
    if mean <= .75:
        return 'lower'
    else:
        return 'higher'

df['primary_dx_tier'] = df['primary_diagnosis'].map(means).apply(tier)
```

```python
df['primary_dx_tier'].value_counts()
```

```python
df2 = df.copy()
df2 = df2.drop(columns=['primary_diagnosis'])
```

```python
df2.to_pickle('cleaned_three.pkl')
```
